# Minimal model
## From Cole et al. (preprint), https://doi.org/10.1101/292045

## Illustrating effect of task timing as confound for task-state FC estimation

### Also: illustrating remaining effects of task timing on time series after "controlling" for task timing, due to temporal autocorrelation (fMRI hemodynamic lag and similar shape across regions)

```python
#Importing packages
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#Setting up task timing
taskDur=150
taskTiming=np.zeros(taskDur).astype('bool')
blockDur=30
#Add 2 blocks to task timing
taskTiming[(blockDur):(2*blockDur)]=1
taskTiming[(3*blockDur):(4*blockDur)]=1

plt.figure(figsize=(5,4))
plt.plot(taskTiming)
plt.xlabel("Time (seconds)")
plt.ylabel("Activity")
plt.title("Task timing")

plt.savefig('minimalmodel_tasktiming.pdf',dpi=600,transparent=True, bbox_inches='tight')
```

```python
#Simulating neural time series without stimuli and without autocorrelation

np.random.seed(83)
regionA_TS_NoStim_NoAutocorr=.25*np.random.randn(taskDur)
np.random.seed(10)
regionB_TS_NoStim_NoAutocorr=.25*np.random.randn(taskDur)

def plotTwoNodeTS(regionAData, regionBData, plottitle, savefilename):
    plt.figure(figsize=(5,4))
    plt.plot(regionAData, label='Region A')
    plt.plot(regionBData, label='Region B')
    plt.xlabel("Time (seconds)")
    plt.ylabel("Activity")
    plt.title(plottitle)
    plt.legend()
    plt.savefig(savefilename,dpi=600,transparent=True, bbox_inches='tight')

plotTwoNodeTS(regionA_TS_NoStim_NoAutocorr,regionB_TS_NoStim_NoAutocorr,
              plottitle="No task stimuli",
              savefilename="minimalmodel_panelA_notaskstim.pdf")

print 'NoStim_NoAutocorr, Correlation between time series, r=', np.corrcoef(regionA_TS_NoStim_NoAutocorr,regionB_TS_NoStim_NoAutocorr)[1,0]
```

```python
#Simulate neural time series with stimuli and without autocorrelation

regionA_TS_Stim_NoAutocorr=regionA_TS_NoStim_NoAutocorr+taskTiming
regionB_TS_Stim_NoAutocorr=regionB_TS_NoStim_NoAutocorr+taskTiming

plotTwoNodeTS(regionA_TS_Stim_NoAutocorr,regionB_TS_Stim_NoAutocorr,
              plottitle="With task stimuli",
              savefilename="minimalmodel_panelB_withtaskstim.pdf")

print 'Stim_NoAutocorr, Correlation between time series, r=', np.corrcoef(regionA_TS_Stim_NoAutocorr,regionB_TS_Stim_NoAutocorr)[1,0]
```

```python
#Isolate task state

regionA_TS_Stim_NoAutocorr_TaskOnly=regionA_TS_Stim_NoAutocorr[taskTiming]
regionB_TS_Stim_NoAutocorr_TaskOnly=regionB_TS_Stim_NoAutocorr[taskTiming]

plotTwoNodeTS(regionA_TS_Stim_NoAutocorr_TaskOnly,regionB_TS_Stim_NoAutocorr_TaskOnly,
              plottitle="With task stimuli, task time points isolated",
              savefilename="minimalmodel_panelC_taskstim_tasktimepointsonly.pdf")

print 'Isolate task state (control for state transitions), Stim_NoAutocorr; Correlation between time series r=', np.corrcoef(regionA_TS_Stim_NoAutocorr_TaskOnly,regionB_TS_Stim_NoAutocorr_TaskOnly)[1,0]
```

```python
#Simulate fMRI data collection
#Creating a hemodynamic response function
#Based on http://www.jarrodmillman.com/rcsds/lectures/convolution_background.html

from scipy.stats import gamma
def hrf(times,peakscale=0.6,peaktime=6,undershoottime=12,undershootratio=.35):
    """ Return values for HRF at given times """
    # Gamma pdf for the peak
    peak_values = gamma.pdf(times, peaktime)
    # Gamma pdf for the undershoot
    undershoot_values = gamma.pdf(times, undershoottime)
    # Combine them
    values = peak_values - undershootratio * undershoot_values
    # Scale max to 0.6
    return values / np.max(values) * peakscale

simsample_rate=1 #Every 1 second
simsample_times = np.arange(0, 30, simsample_rate)
hrf_at_simsample = hrf(simsample_times,0.6,6,12,.35)

plt.figure(figsize=(5,4))
plt.title('fMRI hemodynamic response function')
plt.xlabel("Time (seconds)")
plt.ylabel("Activity")
plt.plot(simsample_times, hrf_at_simsample,label='Normal HRF')

plt.savefig('minimalmodel_HRFshape.pdf',dpi=600,transparent=True, bbox_inches='tight')
```

```python
#Add temporal autocorrelation and common temporal transformation shape (fMRI HRF) via convolution

regionA_TS_NoStim_Autocorr=np.convolve(regionA_TS_NoStim_NoAutocorr, hrf_at_simsample)
n_to_remove = len(hrf_at_simsample) - 1
regionA_TS_NoStim_Autocorr = regionA_TS_NoStim_Autocorr[:-n_to_remove]

regionB_TS_NoStim_Autocorr=np.convolve(regionB_TS_NoStim_NoAutocorr, hrf_at_simsample)
n_to_remove = len(hrf_at_simsample) - 1
regionB_TS_NoStim_Autocorr = regionB_TS_NoStim_Autocorr[:-n_to_remove]

plotTwoNodeTS(regionA_TS_NoStim_Autocorr,regionB_TS_NoStim_Autocorr,
              plottitle="No task stimuli, fMRI HRF convolved",
              savefilename="minimalmodel_paneld_notaskstim_fMRIHRFconvolved.pdf")

print 'NoStim_Autocorr, Correlation between time series, r=', np.corrcoef(regionA_TS_NoStim_Autocorr,regionB_TS_NoStim_Autocorr)[1,0]
```

```python
#Add task events, along with HRF convolution

regionA_TS_Stim_Autocorr=np.convolve(regionA_TS_Stim_NoAutocorr, hrf_at_simsample)
n_to_remove = len(hrf_at_simsample) - 1
regionA_TS_Stim_Autocorr = regionA_TS_Stim_Autocorr[:-n_to_remove]

regionB_TS_Stim_Autocorr=np.convolve(regionB_TS_Stim_NoAutocorr, hrf_at_simsample)
n_to_remove = len(hrf_at_simsample) - 1
regionB_TS_Stim_Autocorr = regionB_TS_Stim_Autocorr[:-n_to_remove]

plotTwoNodeTS(regionA_TS_Stim_Autocorr,regionB_TS_Stim_Autocorr,
              plottitle="With task stimuli, fMRI HRF convolved",
              savefilename="minimalmodel_panelF_taskstim_fMRIHRFconvolved.pdf")

print 'Stim_Autocorr, Correlation between time series, r=', np.corrcoef(regionA_TS_Stim_Autocorr,regionB_TS_Stim_Autocorr)[1,0]
```

```python
#Isolate task state

#Increase event duration to account for temporal autocorrelation (extended HRF shape)
tasktiming_fMRI=np.zeros(taskDur).astype('bool')
tasktiming_fMRI[(blockDur):(2*blockDur)]=1
tasktiming_fMRI[(2*blockDur):(2*blockDur+10)]=1
tasktiming_fMRI[(3*blockDur):(4*blockDur)]=1
tasktiming_fMRI[(4*blockDur):(4*blockDur+10)]=1

regionA_TS_Stim_Autocorr_TaskOnly=regionA_TS_Stim_Autocorr[tasktiming_fMRI]
regionB_TS_Stim_Autocorr_TaskOnly=regionB_TS_Stim_Autocorr[tasktiming_fMRI]

plotTwoNodeTS(regionA_TS_Stim_Autocorr_TaskOnly,regionB_TS_Stim_Autocorr_TaskOnly,
              plottitle="fMRI HRF convolved, task time points isolated",
              savefilename="minimalmodel_panelG_taskstim_fMRIHRFconvolved_taskonly.pdf")

print 'Isolate task state (control for state transitions), Stim_Autocorr; Correlation between time series r=', np.corrcoef(regionA_TS_Stim_Autocorr_TaskOnly,regionB_TS_Stim_Autocorr_TaskOnly)[1,0]
```

```python
#Try to estimate "steady state" during task blocks (reducing effect of start and stop transition transients)

tasktiming_restricted=np.zeros(taskDur).astype('bool')
tasktiming_restricted[(blockDur+10):(2*blockDur)]=1
tasktiming_restricted[(3*blockDur+10):(4*blockDur)]=1

regionA_TS_Stim_Autocorr_restricted=regionA_TS_Stim_Autocorr[tasktiming_restricted]
regionB_TS_Stim_Autocorr_restricted=regionB_TS_Stim_Autocorr[tasktiming_restricted]

plotTwoNodeTS(regionA_TS_Stim_Autocorr_restricted,regionB_TS_Stim_Autocorr_restricted,
              plottitle="fMRI HRF convolved, main task 'transients' removed",
              savefilename="minimalmodel_panelH_taskstim_fMRIHRFconvolved_notasktransients.pdf")

print 'Stim_Autocorr, transitions cut off; Correlation between time series r=', np.corrcoef(regionA_TS_Stim_Autocorr_restricted,regionB_TS_Stim_Autocorr_restricted)[1,0]
```

```python
#Calculate mean evoked response

event1timing=np.zeros(taskDur).astype('bool')
event1timing[(blockDur):(2*blockDur)]=1
#Increase event duration to account for temporal autocorrelation (extended HRF shape)
event1timing[(2*blockDur):(2*blockDur+20)]=1
event2timing=np.zeros(taskDur).astype('bool')
event2timing[(3*blockDur):(4*blockDur)]=1
#Increase event duration to account for temporal autocorrelation (extended HRF shape)
event2timing[(4*blockDur):(4*blockDur+20)]=1

regionA_meanEvokedResponse=(regionA_TS_Stim_Autocorr[event1timing]+regionA_TS_Stim_Autocorr[event2timing])/2
regionB_meanEvokedResponse=(regionB_TS_Stim_Autocorr[event1timing]+regionB_TS_Stim_Autocorr[event2timing])/2

plotTwoNodeTS(regionA_meanEvokedResponse,regionB_meanEvokedResponse,
              plottitle="Mean evoked responses",
              savefilename="minimalmodel_meanevokedresponses.pdf")

print 'Mean evoked responses; Correlation between time series r=', np.corrcoef(regionA_meanEvokedResponse,regionB_meanEvokedResponse)[1,0]
```

```python
#Subtract mean evoked responses

regionA_TS_Stim_Autocorr_meanSubt=regionA_TS_Stim_Autocorr.copy()
regionA_TS_Stim_Autocorr_meanSubt[event1timing]=regionA_TS_Stim_Autocorr_meanSubt[event1timing]-regionA_meanEvokedResponse
regionA_TS_Stim_Autocorr_meanSubt[event2timing]=regionA_TS_Stim_Autocorr_meanSubt[event2timing]-regionA_meanEvokedResponse
regionA_TS_Stim_Autocorr_meanSubt=regionA_TS_Stim_Autocorr_meanSubt[tasktiming_fMRI]

regionB_TS_Stim_Autocorr_meanSubt=regionB_TS_Stim_Autocorr.copy()
regionB_TS_Stim_Autocorr_meanSubt[event1timing]=regionB_TS_Stim_Autocorr_meanSubt[event1timing]-regionB_meanEvokedResponse
regionB_TS_Stim_Autocorr_meanSubt[event2timing]=regionB_TS_Stim_Autocorr_meanSubt[event2timing]-regionB_meanEvokedResponse
regionB_TS_Stim_Autocorr_meanSubt=regionB_TS_Stim_Autocorr_meanSubt[tasktiming_fMRI]

plotTwoNodeTS(regionA_TS_Stim_Autocorr_meanSubt,regionB_TS_Stim_Autocorr_meanSubt,
              plottitle="fMRI HRF convolved, mean evoked responses subtracted",
              savefilename="minimalmodel_panelI_fMRIwithmeanevokedresponsessubt.pdf")

print 'Removing mean evoked responses, Stim_Autocorr; Correlation between time series r=', np.corrcoef(regionA_TS_Stim_Autocorr_meanSubt,regionB_TS_Stim_Autocorr_meanSubt)[1,0]
```
